```python
# 6. Explore the top-selling products and countries based on the quantity sold.
```

```python
import pandas as pd
import matplotlib.pyplot as plt
top_products = df.groupby("Description")["Quantity"].sum().sort_values(ascending=False).head(10)
top_countries = df.groupby("Country")["Quantity"].sum().sort_values(ascending=False).head(10)
```

```python
df = pd.read_csv("OnlineRetail(in).csv")
```

```python
# Plot Top Selling Products
plt.figure(figsize=(12, 6))
top_products.plot(kind='bar', color='skyblue')
plt.title("Top 10 Selling Products by Quantity")
plt.xlabel("Product Description")
plt.ylabel("Total Quantity Sold")
plt.xticks(rotation=90)
plt.show()
```

```python
# Plot Top Selling Countries
plt.figure(figsize=(12, 6))
top_countries.plot(kind='bar', color='lightcoral')
plt.title("Top 10 Selling Countries by Quantity")
plt.xlabel("Country")
plt.ylabel("Total Quantity Sold")
plt.xticks(rotation=45)
plt.show()
```
